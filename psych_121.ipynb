## **Finish Time** 


```python
# convert the StartDate and EndDate columns to datetime objects
psych['StartDate'] = pd.to_datetime(psych['StartDate'])
psych['EndDate'] = pd.to_datetime(psych['EndDate'])

# calculate the difference between the StartDate and EndDate columns and store the result in a new column
psych['TotalDate'] = (psych['EndDate'] - psych['StartDate']).astype('timedelta64[s]') / 60

#Average time
total_date = psych['TotalDate'].mean()
total_date_rounded = round(total_date,2)
print ("Average time of completion of survery:  ", total_date_rounded)
```

```python
filtered = psych[psych['TotalDate'] < 7]

# Calculate the percentage of filtered values
percentage = (len(filtered) / len(psych)) * 100
percentage

print("Percentage of people who got a score lower than 7:", percentage)

```

## **Finding Count of the Words**


```python
#Number of Words
text_column = psych['Black']

# Define a function to count the number of words in a string
def count_words(text):
    return len(text.split())

# Apply the function to the text column
word_counts = text_column.apply(count_words)

# Replace the words with the number of words
psych['Black'] = word_counts
```

```python
#Number of Words
text_column = psych['Red']

# Define a function to count the number of words in a string
def count_words(text):
    return len(text.split())

# Apply the function to the text column
word_counts = text_column.apply(count_words)

# Replace the words with the number of words
psych['Red'] = word_counts
```

<p> A survey created on Qualtrics was distributed to college students and faculty from the University of California, Los Angeles. The target population was young college students and faculty who have not been diagnosed with any eye condition or disease, don't have color blindness and are not taking medication that affects vision. In addition to blind convenience sampling in UCLA GroupMe group chats, the recruitment strategy involved directly approaching the author's friends and requesting them to distribute the survey after finishing it. Out of the 22 participants, all were shown the Black Condition, Red Condition, and the Green Condition. </p>
